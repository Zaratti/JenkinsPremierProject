#Visualizing the Trend of Children Under Age 5 with Fever Over the Years

```python
# Plotting the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='% of children under age 5 with fever', data=df, marker='o')
plt.title('Trend of Children Under Age 5 with Fever Over Years')
plt.xlabel('Year')
plt.ylabel('% of Children Under Age 5 with Fever')
plt.show()


# conclusion
# The chart illustrates a decrease in the percentage of children under 5 years old who have fever over the years. This trend is likely attributable to advancements in the medical field.
```

#The chart illustrates a decrease in the percentage of children under 5 years old who have fever over the years. This trend is likely attributable to advancements in the medical field.

```python
# Select the numerical columns
numerical_columns = ['Rural', 'Urban', 'Poorest', 'Second', 'Middle', 'Fourth', 'Richest']

plt.figure(figsize=(15, 10))

# Plot each numerical column on the same graph
for col in numerical_columns:
    sns.lineplot(data=region_data, x=region_data.index, y=col, label=col)

# Adding title and labels
plt.title('Line Plot of Numerical Columns')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend(title='Columns')

plt.show()
```

# We observe that in the Middle East and North Africa, the likelihood of having malaria is relatively high regardless of socioeconomic status, whether rich or poor, and irrespective of living in rural areas. Conversely, in Eastern and Southern Africa, as well as West and Central Africa, the likelihood of contracting malaria is significantly lower.



```python
# Select the numerical columns
numerical_columns = ['None',	'Primary',	'Sec & Higher']

plt.figure(figsize=(15, 10))

# Plot each numerical column on the same graph
for col in numerical_columns:
    sns.lineplot(data=region_data, x=region_data.index, y=col, label=col)

# Adding title and labels
plt.title('Line Plot of Numerical Columns')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend(title='Columns')

plt.show()
```

#From the chart above, we can see that the Middle East and North Africa have the least health awareness and practices, contributing to higher malaria prevalence and increased drug needs. This may explain the greater likelihood of malaria in these regions. In contrast, areas like Sub-Saharan Africa and West Africa exhibit higher health awareness and better practices for malaria prevention, resulting in a lower likelihood of malaria incidence.

#Histograms for selected variables (e.g., population growth, percentage of children under age 5 with fever)

```python
data = {'Category': ['A', 'B', 'C', 'D'],
        'Values': [10, 20, 30, 40]}
#Distribution of Selected Variables e.g  population growth,% of children under age 5 with fever etc to the histograms

descriptive_stats = df.describe()

# Visualization - Histograms for selected columns
import matplotlib.pyplot as plt

def plot_histograms(dataframe):
    plt.figure(figsize=(12, 10))
    for index, column in enumerate(['Population_growth(%)', '% of children under age 5 with fever']):
        plt.subplot(2, 2, index + 1)
        dataframe[column].hist(bins=15)
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

descriptive_stats, plot_histograms(df)

```

#Distribution of population_growth(%)- The majority of countries have population growth rate between 1% to 3%, there are few countries with negative or extremely high growth rate.The majority of countries have low percentage(0% to 10) of children under age 5 with fever there is a wide range of percentages with some countries having up to 70% of the childdren under age 5 with fever, the distribution is right skewed indicating a high percentage of children with fever are less common but do exist.

#Bar Plot of Children Under Age 5 with Fever by World Bank Income Group

```python
data = {'Category': ['Lower middle income', 'Low income', "Upper middle income"],
        'Values': [20, 23, 13]}  # Using integers for percentages

df = pd.DataFrame(data)

# Plotting the bar chart
plt.figure(figsize=(14, 7))
sns.barplot(x='Category', y='Values', data=df)
plt.title('Children Under Age 5 with Fever by Income Group')
plt.xlabel('World Bank Income Group (2022)')
plt.ylabel('% of Children Under Age 5 with Fever')

# Adding actual percentage values as text labels on top of each bar
for index, value in enumerate(df['Values']):
    plt.text(index, value + 0.5, f'{value}%', ha='center', va='bottom', fontsize=10)

plt.show()
#conclusion is that the data indicates that lower income groups are significantly more likely to have malaria or children with malaria compared to higher income groups. This suggests a strong correlation between income level and malaria prevalence, with lower income earners being disproportionately affected.
```

#The data indicates that lower income groups are significantly more likely to have malaria or children with malaria compared to higher income groups. This suggests a strong correlation between income level and malaria prevalence, with lower income earners being disproportionately affected.

#Heatmap of Malaria Distribution Across Countries and Years

```python
# Assuming df is your DataFrame with duplicate entries
# Drop duplicate rows
df_no_duplicates = df.drop_duplicates(subset=['Countries and areas', 'Year'])

# Create a pivot table with mean aggregation
pivot_table = df_no_duplicates.pivot_table(index='Countries and areas', columns='Year',
                                           values='% of children under age 5 with fever', aggfunc='mean')

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data=pivot_table, cmap='viridis')
plt.title("Heatmap of Malaria Distribution")
plt.xlabel("Year")
plt.ylabel("Countries and areas")
plt.show()
```

#The place were malaria are prevailent to be darker and places where they are not as prevailent to be lighter.

 #Bar plot showing the percentage of children under age 5 with fever for different demographic groups and saves the plot as a PNG image.



```python
demographics_mode_of_living = ['Rural', 'Urban', 'Poorest', 'Richest']
df_melted = df.melt(id_vars=['Year'], value_vars=demographics_mode_of_living, var_name='Demographic', value_name='Percentage')

# Define the demographic categories and their corresponding percentages
data = {'Category': ['Rural', 'Urban', 'Poorest', 'Richest'],
        'Percentage': [35, 35, 33, 35]}

# Set the figure size
plt.figure(figsize=(8, 6))

# Create a bar plot using the data dictionary directly
sns.barplot(x='Category', y='Percentage', data=data)

# Add title and labels to the plot
plt.title('Percentage of Children Under Age 5 with Fever by Demographic Group')
plt.xlabel('Demographic Group')
plt.ylabel('Percentage')

# Add percentage values as text labels on top of each bar
for index, value in enumerate(data['Percentage']):
    plt.text(index, value + 0.5, f'{value}%', ha='center', va='bottom', fontsize=10)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Adjust layout for better appearance
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.25)

# Save the plot as a PNG image with a specific size and then display it
plt.savefig('bar_plot.png', dpi=300)
plt.close()
plt.show()

```
